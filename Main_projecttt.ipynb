```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/most_runs_average_strikerate.csv')

top10 = df.nlargest(10, 'average')
plt.bar(top10.batsman, top10.average, color='gold', edgecolor='black')
plt.title('Does Survival Cause a Higher Average?')
plt.xticks(rotation=45)
plt.show()

ind_sum = df[df.batsman.isin(['V Kohli', 'SK Raina', 'RG Sharma', 'S Dhawan', 'MS Dhoni'])].total_runs.sum()
row_sum = df[df.batsman.isin(['DA Warner', 'CH Gayle', 'AB de Villiers', 'SR Watson', 'KA Pollard'])].total_runs.sum()
eng_sum = df[df.batsman.isin(['JC Buttler', 'KP Pietersen', 'EJG Morgan', 'BA Stokes', 'JJ Roy'])].total_runs.sum()

plt.bar(['India', 'Overseas'], [ind_sum, row_sum], color=['black', 'white'], edgecolor='royalblue')
plt.title('Improvement Over Time')
plt.show()

plt.scatter(df.total_runs, df.out, c=df.average)
plt.title('The Bad Thing: Increased Risk')
plt.xlabel('Total Runs')
plt.ylabel('Times Out')
plt.colorbar(label='Average')
plt.show()
```
